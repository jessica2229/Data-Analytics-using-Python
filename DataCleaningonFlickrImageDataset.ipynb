```python
pub = df['Place of Publication']
london = pub.str.contains('London')
london[:5]
```

```python
oxford = pub.str.contains('Oxford')
df['Place of Publication'] = np.where(london, 'London',np.where(oxford, 'Oxford', pub.str.replace('-', ' ')))
df['Place of Publication'].head()
```

```python
towns_df = pd.DataFrame(df, columns=['Title'])
towns_df.head()
```

```python
def get_title(item):
    if ' (' in item:
        return item[:item.find(' (')]
    elif '[' in item:
        return item[:item.find('[')]
    else:
        return item
```

```python
towns_df = towns_df.applymap(get_title)
towns_df.head(15)
```

```python
def get_title(item):
    if 'By' in item:
        return item[:item.find('By')]
    else:
        return item
```

```python
towns_df = towns_df.applymap(get_title)
towns_df.head(15)
```
